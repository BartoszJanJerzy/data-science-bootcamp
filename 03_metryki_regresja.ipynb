```
def regression_plot(y_true,y_pred):
  df = pd.DataFrame({'y_true':y_true,'y_pred':y_pred,'pred_error':y_pred-y_true})
  
  #wartości min i maks dla wyników prawdziwych
  min = results['y_true'].min()
  max = results['y_true'].max()

  fig = go.Figure(data=[go.Scatter(
                            x=df['y_true'],
                            y=df['y_pred'],
                            name='Predicted results',
                            mode='markers'
                        ),
                        go.Scatter(
                            x=y_true,
                            y=y_true,
                            mode='lines',
                              line_dash='dash',
                            name='Real results'
                        )],
                       layout=go.Layout(
                           title='Regression results',
                           showlegend=False,
                           xaxis_title='y_true',
                           yaxis_title='y_pred'
                           )
                       )
  fig.show()
  
```

## Mean Absolute Error - Średni błąd bezwzględny (MAE)

```
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_true,y_pred)

# a ręcznie
def MeanAbsoluteError(y_true,y_pred):
  return abs(y_true-y_pred).sum() / len(y_true)

mae2 = MeanAbsoluteError(y_true,y_pred)


print(mae,mae2,sep='\n')
```

## Mean Squared Error - MSE - Błąd średniokwadratowy
(przez podneisienie do kwadratu sumy błędów, wyniki skrajne są bardziej karane)

```
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_true,y_pred)

# a ręcznie
def MeanSquaredError(y_true,y_pred):
  return ((y_true-y_pred) ** 2).sum() / len(y_true)

mse2 = MeanSquaredError(y_true,y_pred)


print(mse,mse2,sep='\n')
```

## Root Mean Squared Error - RMSE - Pierwiastek błędu średniokwadratowego
(jak średnio nasze przewidziane wartości odbiegają od wartości rzeczywistych)

```
def RootMeanSquaredError(y_true,y_pred):
  return np.sqrt(((y_true-y_pred) ** 2).sum() / len(y_true))

rmse = RootMeanSquaredError(y_true,y_pred)
rmse
```

## Max Error - błąd maksymalny

```
from sklearn.metrics import max_error
me = max_error(y_true,y_pred)

# a ręcznie
def MaxError(y_true,y_pred):
  return abs(y_true - y_pred).max()

me2 = MaxError(y_true,y_pred)


print(me,me2,sep='\n')
```

## R2 score - współczynnik determinacji
(o ile lepiej nasz model przewiduej dane, niż model przypisujący po prostu średnią z całości)

```
from sklearn.metrics import r2_score
r2_score = r2_score(y_true,y_pred)

# a ręcznie
def R2(y_true,y_pred):
  numerator = ((y_true-y_pred)**2).sum()
  denominator = ((y_true-y_true.mean())**2).sum()

  try:
    r_squared = 1 - numerator/denominator
  except ZeroDivisionError:
    print('Dzielenie przez zero')
  
  return r_squared

r_squared = R2(y_true,y_pred)

print(r2_score,r_squared,sep='\n')
```
