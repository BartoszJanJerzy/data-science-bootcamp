## Siatka obrazów

```
fig = plt.figure(figsize=(12,10))

plt.subplot(221)
plt.imshow(mpimg.imread('Fotka dla Miesia 2 BW.JPG'))
plt.subplot(222)
plt.imshow(mpimg.imread('Fotka dla Miesia 2 BW2.JPG'))
plt.subplot(223)
plt.imshow(mpimg.imread('Fotka dla Miesia BW.JPG'))
plt.subplot(224)
plt.imshow(mpimg.imread('Fotka dla Miesia BW2.JPG'))
```

```
# a jak to zautomatyzować?
fig = plt.figure(figsize=(12,10))

photos = ['Fotka dla Miesia 2 BW.JPG','Fotka dla Miesia BW2.JPG','Fotka dla Miesia BW.JPG','Fotka dla Miesia 2 BW2.JPG']
for idx, img in enumerate(photos):
  plt.subplot(220+idx+1)
  plt.imshow(mpimg.imread(img))
```

```
# jak zmienić rozmiar siatki na 1 wiersz?

for idx, img in enumerate(photos):
  plt.subplot(140+idx+1)
  plt.imshow(mpimg.imread(img))
  plt.axis('Off')
```
