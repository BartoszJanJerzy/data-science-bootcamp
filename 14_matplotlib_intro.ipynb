```
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```
import seaborn as sns
sns.set()
```

## Najprostszy wykres liniowy

```
# najprostszy wykres z dwoma punktami
_ = plt.plot([0,1],[0,3])
```

```
# modyfikacja wykresu
plt.plot([0,1,2],[0,3,2], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend() # czy jest legenda
```

## Wykres sinus()/cos()

```
# najporściej
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)

_ = plt.plot(x, x_sin)
```

```
# a wiecej wykresów na jednym obiekcie
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

## Tworzenie wielu wykresów

```
# na początku...
plt.subplot(121)
_ = plt.plot([0,1,2],[3,2,5])

# 1 - ile wierszy z wykresami
# 2 - ile kolumn z wykresami
# 1 - odniesienie się do konkretneg wykresu (pierwsyz z lewej)
```

```
# a jak dodać drugi i trzeci wykres?
plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,1,4], 'r--')
plt.subplot(133)
plt.plot([0,1,2],[3,1,4], 'ro')
```

```
# zmiana rozmiaru
fig = plt.figure(figsize=(14,8))

plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,1,4], 'r--')
plt.subplot(133)
plt.plot([0,1,2],[3,1,4], 'ro')
```

## Wykres słupkowy

```
plt.bar(x=['a','b','c','d'], height=[1.2,3.2,1.2,5.2])
```

```
# a poziomo
plt.barh(y=['a','b','c','d'], width=[1.2,3.2,1.2,5.2])
```

## Wykres punktowy

```
x = np.random.randn(300)
y = np.random.randn(300)

plt.scatter(x,y)
```

```
# przykładowa modyfikacja punktów
size = np.random.randn(300) * 50

_ = plt.scatter(x,y,size, alpha = 0.5) #alpha -> przezroczystość
```
